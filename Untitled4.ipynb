```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image=cv2.imread('/content/sample_data/obinabi.jpg')
(h,w)=image.shape[:2]
center=(w//2,h//2)
angle = 180
scale = 1.0
rotation_matrix = cv2.getRotationMatrix2D(center,angle,scale)
rotated = cv2.warpAffine(image, rotation_matrix,(w, h))
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.title("Rotated Image(180 degree)")
plt.show()
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from IPython.display import clear_output
import time
video_path = '/content/sample_data/neymar.html'
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()
fps = cap.get(cv2.CAP_PROP_FPS)
delay = 10/fps
while cap.isOpened():
  ret, frame = cap.read()
  if not ret:
    print("end of video")
    break
    clear_output(wait=True)
  cv2_imshow(frame)
  time.sleep(delay)
cap.release()
cv2.destroyAllWindows()
```
